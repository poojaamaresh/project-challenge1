```python
# importing libraries
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score



# loading the training dataset
df = pd.read_csv('wine.csv')
# Replace the target variable with bad-0, good 1
df['quality'].replace(['good', 'bad'],[1, 0], inplace=True)
df=df.assign(bias=1599*[1])
# shape of the dataframe
df.shape
#print(df.head())
df.head()
# defining the dependent and independent variables

#Its mandatory to add consttant to get the coefficient

#allocate first 1400 rows as training data
#ytrain = 
#Xtrain = 
# building the model and fitting the data
#log_reg = 
```

```python
y = df['quality']
x=df.loc[:, df.columns!='quality']
x.head()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,random_state=0,train_size=.87)
# Test data from row number 1401 to the end
#Xtest = 
#ytest = 


# performing predictions on the test datdaset
#yhat = 
#prediction = logit_model.predict([,2.62,1])

# comparing original and predicted values of y
#print('Actual values', list(ytest.values))
#print('Predictions :', prediction)
```

```python
logit_model=sm.Logit(y_train,x_train).fit()
print(logit_model.summary())
```

```python
# import required libraries to output confusion matrix and accuracy score
y_pred=logit_model.predict(x_test)

y_pred=round(y_pred)
y_pred
# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, y_pred))
```
