### PCA as visualization tool
**We will visualize the data along first two principle components**

Towards his we consider a classifiaction probelsem.
We use wine data set that has 13 features and we canot visualize in 
13 dimension. So using first two principle components we see the data.
Apart from this we also see why scaling is important.


```python
# Load wine data set and use PCA as visulaization tool
# and plot the classes
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#Libraries
```

```python
#read the data
#winedata=pd.read_csv('wine.csv')
# Load dataset
winedata = load_wine()
X, y = winedata['data'] , winedata['target']
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X[0]
```

```python
#categories=np.array(winedata)
#categories
#colormap=np.array(['r','g','b'])
#colormap
load_wine
```

```python
#use scatter plot and show the target variable using 2 features
categories=np.array(y)
colormap=np.array(['r','g','b'])
plt.scatter(X[:,0], X[:,1], s=30,c=colormap[categories], marker='o')
#plt.plot(X[:,0], X[:,1], color='tab:blue')
plt.xlabel(winedata["feature_names"][1])
plt.ylabel(winedata["feature_names"][2])
plt.title("Two particular features of the wine dataset")
plt.plot()
```

```python
#sc = StandardScaler()
#X_train = sc.fit_transform(X)
#X_test = sc.transform(X_test)
```

```python
pca = PCA()
 
Xt = pca.fit_transform(X)
```

```python
plot=plt.scatter(Xt[:,0], Xt[:,1] , c=y)
plt.legend(handles=plot.legend_elements()[0], labels=list(winedata['target_names']))
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("First two principal components")
plt.plot()
```

```python
#fit_transform 
sc = StandardScaler()
Xsc = sc.fit_transform(X)
Xscpca = pca.fit_transform(Xsc)
```

```python
plot=plt.scatter(Xscpca[:,0], Xscpca[:,1] , c=y)
plt.xlabel("SCPC1")
plt.ylabel("SCPC2")
plt.title("First two principal components")
plt.plot()
```

```python
#use scatter plot and show the target variable using 3 features
x_train, x_test, y_train, y_test= train_test_split(Xscpca,y,random_state=0,train_size=.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)

# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, y_pred))
```
